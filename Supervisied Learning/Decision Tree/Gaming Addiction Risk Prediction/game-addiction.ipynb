```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Step 1: Create the dataset
data = {
    'Hours_Gaming_Per_Week': [5, 10, 25, 40, 35, 50, 20, 15, 60, 55, 12, 8, 30, 45, 22],
    'Sleep_Hours': [8, 7, 6, 5, 5, 4, 7, 6, 4, 4, 7, 8, 6, 5, 7],
    'Social_Activity_Score': [8, 7, 4, 2, 3, 1, 5, 6, 1, 2, 7, 8, 4, 2, 5],
    'Academic_Performance_Score': [9, 8, 6, 3, 4, 2, 6, 7, 2, 3, 8, 9, 5, 3, 6],
    'Risk': ['Low', 'Low', 'High', 'High', 'High', 'High', 'Low', 'Low', 'High', 'High', 'Low', 'Low', 'High', 'High', 'Low']
}

df = pd.DataFrame(data)
```

```python
# Step 2: Encode the target variable
le = LabelEncoder()
df['Risk'] = le.fit_transform(df['Risk'])  # Low = 0, High = 1
```

```python
# Step 3: Split features and target
X = df.drop('Risk', axis=1)
y = df['Risk']
```

```python
# Step 4: Train the Decision Tree model
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)
```

```python
# Step 5: Make predictions and evaluate
y_pred = model.predict(X)
accuracy = accuracy_score(y, y_pred)
report = classification_report(y, y_pred, target_names=['Low', 'High'])

print("Accuracy:", accuracy)
print("\nClassification Report:\n", report)
```

```python
# Step 6: Visualize the Decision Tree
plt.figure(figsize=(12, 8))
tree.plot_tree(model, feature_names=X.columns, class_names=['Low', 'High'], filled=True)
plt.title("Decision Tree - Gaming Addiction Risk Prediction")
plt.show()
```

```python
#| Feature                      | Description                                      |
#| ---------------------------- | ------------------------------------------------ |
#| `Hours_Gaming_Per_Week`      | Time spent gaming per week (numeric)             |
#| `Sleep_Hours`                | Average sleep hours per day                      |
#| `Social_Activity_Score`      | Score out of 10 for social activity              |
#| `Academic_Performance_Score` | Score out of 10 for academic/work performance    |
#| `Risk`                       | Target label: `'Low'` or `'High'` addiction risk |
```
