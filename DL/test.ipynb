```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate some fake data: y = 3x + 2 + noise
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 samples
y = 3 * X + 2 + np.random.randn(100, 1)

# 2. Initialize parameters
w = np.random.randn()
b = np.random.randn()

# Hyperparameters
lr = 0.1  # learning rate
epochs = 100

# 3. Training loop (gradient descent)
for epoch in range(epochs):
    # Predictions
    y_pred = w * X + b

    # Compute loss (Mean Squared Error)
    loss = np.mean((y_pred - y) ** 2)

    # Compute gradients
    dw = np.mean(2 * (y_pred - y) * X)
    db = np.mean(2 * (y_pred - y))

    # Update parameters
    w -= lr * dw
    b -= lr * db

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: loss={loss:.4f}, w={w:.2f}, b={b:.2f}")

# 4. Plot result
plt.scatter(X, y, color="blue")
plt.plot(X, w*X + b, color="red")
plt.xlabel("x (house size)")
plt.ylabel("y (price)")
plt.show()
```
