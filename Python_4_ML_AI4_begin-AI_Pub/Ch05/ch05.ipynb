```python
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals=np.linspace(0,20,20)
y_vals=[math.sqrt(i) for i in x_vals]

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Square Roots')
plt.plot(x_vals, y_vals, 'y')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals=np.linspace(0,20,20)
y_vals=[math.sqrt(i) for i in x_vals]

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Square Roots')
plt.plot(x_vals, y_vals, 'y',label='Square Root')
plt.legend(loc='upper center')
```
